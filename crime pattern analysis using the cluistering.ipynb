1.  Standardizing the "STATE/UT" COLUMNS

The "State/UT" columns in the dataset contains state names in both uppercase and lower case format. To maintain consistency across the dataset, all entries iin this column are converted to upper case .This ensures that state names are standardized and prevents any inconsistencies when analyze the data


2.   Removing  Invalid disticts Entries

The "distict" column includes some entries like "TOTAL" and "ZZ TOTAL ",which are not valid distict names but are being considered as such in the dataset. These entries represent aggregates and not specific disticts ,so they are renmoved from the dataset to ensures that only valid distict data is retained .removing these entries helps in maintaining the integrity of the analysis

```python
import pandas as pd
```

```python
url="D:/career/project/crime data/crime.csv"
data=pd.read_csv(url)
data["STATE/UT"]=data["STATE/UT"].str.upper()
```

```python
if "DISTRICT" in data.columns:
    data=data[~data["DISTRICT"].str.upper().isin(["TOTAL","ZZ TOTAL"])]
else:
    print("column 'DISTRICT' not found .check for spelling or formatting mistake") 
output_path="crimes_cleaned2.csv"
data.to_csv(output_path,index=False)
print("Data processing completed. modified file saved to",output_path)
```
