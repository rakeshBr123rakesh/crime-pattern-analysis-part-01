1.  Standardizing the "STATE/UT" COLUMNS

The "State/UT" columns in the dataset contains state names in both uppercase and lower case format. To maintain consistency across the dataset, all entries iin this column are converted to upper case .This ensures that state names are standardized and prevents any inconsistencies when analyze the data


2.   Removing  Invalid disticts Entries

The "distict" column includes some entries like "TOTAL" and "ZZ TOTAL ",which are not valid distict names but are being considered as such in the dataset. These entries represent aggregates and not specific disticts ,so they are renmoved from the dataset to ensures that only valid distict data is retained .removing these entries helps in maintaining the integrity of the analysis

```python
import pandas as pd
```

```python
url="D:/career/project/crime data/crime.csv"
data=pd.read_csv(url)
data["STATE/UT"]=data["STATE/UT"].str.upper()
```

```python
if "DISTRICT" in data.columns:
    data=data[~data["DISTRICT"].str.upper().isin(["TOTAL","ZZ TOTAL"])]
else:
    print("column 'DISTRICT' not found .check for spelling or formatting mistake") 
output_path="crimes_cleaned2.csv"
data.to_csv(output_path,index=False)
print("Data processing completed. modified file saved to",output_path)
```

data visualization 

overview

visualizing the data provides valuable insights into crime tends ,distribution and properties. In this section ,we use box plot,p[ie chart and trends lines to better understanding the patterns and dynamics of crimes in india

```python
columns_of_interest=["MURDER","RAPE","TOTAL IPC CRIMES"]
plt.figure(figsize=(12,8))
for i,col in enumerate (columns_of_interest,1):
    plt.subplot(1,3,i)
    sns.boxplot(y=data[col])
    plt.title(f"Boxplot of {col}")

plt.show()
```

```python
# yearly tend of total IPC crimes 
plt.figure(figsize=(10,6))
data.groupby("YEAR")["TOTAL IPC CRIMES"].sum().plot(marker="o")
plt.title("tends of total IPC Over Years")
plt.xlabel("year")
plt.ylabel("Total IPC Crimes")
plt.xticks(rotation=45,ha="right")
plt.grid(True)
#plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
state_mean=data.groupby("YEAR")["TOTAL IPC CRIMES"].mean()
#Create the pie chart
plt.figure(figsize=(15,8))
plt.pie(state_mean,labels=state_mean.index,startangle=90,shadow=True,
       textprops={"fontsize":10,"color":"green"},autopct="%0.2f%%")
plt.title("CRIME IN INDIA")
plt.show()
```

```python
crime_total=data[["MURDER","RAPE","KIDNAPPING & ABDUCTION","ROBBERY","BURGLARY","THEFT"]].sum()
plt.figure(figsize=(10,8))
plt.pie(crime_total.values,labels=crime_total.index,autopct='%1.1f%%')
plt.title("Proportion of Crimes by Type")
plt.show()
```

```python
#group group data by state/union territory and sum the total number of IPC crimes
state_totals=data.groupby('STATE/UT')['TOTAL IPC CRIMES'].sum()
# Create a bar chart
plt.figure(figsize=(15,10))
plt.bar(state_totals.index,state_totals.values)
plt.xlabel("State/UT")
plt.ylabel("Total IPC crimes")
plt.title("Total Crimes by State/Union Territory")
plt.xticks(rotation=90)
plt.show()
```

```python
# create a scatterplot theft and burglary crimes
plt.figure(figsize=(12,6))
plt.scatter(data["THEFT"],data["BURGLARY"])
plt.xlabel("Total Theft crimes")
plt.ylabel("Total Burglary Crimes")
plt.title("Theft vs Burglary")
plt.show()
```
